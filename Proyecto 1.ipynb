 ###  Elección del conjunto de datos
 
El conjunto de datos seleccionado hace referencia al rendimiento estudiantil, el cual incluye diferentes variables que se consideran en la investigación educativa, las cuales pueden influir en el éxito académico de un estudiante.

El dataset fue descargado de:

https://www.kaggle.com/datasets/lainguyn123/student-performance-factors/data

#### Descripción de Columnas 

* Hours_Studied: Horas dedicadas a estudiar por semana.
* Attendance: Porcentaje de clases asistidas.
* Parental_Involvement: Nivel de involucramiento de los padres (Bajo, Medio, Alto).
* Access_to_Resources: Acceso a recursos educativos (Bajo, Medio, Alto).
* Extracurricular_Activities: Participación en actividades extracurriculares (Si, No).
* Sleep_Hours: Promedio de horas de sueño.
* Previous_Scores: Puntaje de exámenes anteriores.
* Motivation_Level: Nivel de Motivación (Bajo, Medio, Alto).
* Internet_Access: Acceso A internet (Si, No).
* Tutoring_Sessions:  Numero de tutorías Mensuales.
* Family_Income: Nivel de Ingreso Familiar (Bajo, Medio, Alto).
* Teacher_Quality: Calidad de los profesores (Bajo, Medio, Alto).
* School_Type:Tipo de Escuela (Publica, Privada).
* Peer_Influence: Influencia de Compañeros (Positiva, Neutral, Negativa).
* Physical_Activity: Promedio de actividad física por semana.
* Learning_Disabilities: Problema de aprendizaje (Si, No).
* Parental_Education_Level: Nivel educativo Padres (Secundaria, Universitaria, PostGrado). 
* Distance_from_Home: Distancia de casa a la Escuela (Cerca, Moderado, Lejos).
* Gender: Genero.
* Exam_Score: Puntaje Examen Final.

###  Limpieza y transformación de los datos



```python
# validamos valores faltantes
print(df.isnull().sum()) 
```

```python
#Eliminamos valores faltantes 
# Al  Por no ser un porcentaje elevado, se tomó la decisión se eliminarlos  
df = df.dropna()
```

```python
# Validar Valores Duplicados
df.duplicated().sum()
```

```python
# Verificando notas mayores a 100
# En el resumen estadístico nos dimos cuenta que existían valores mayores a 100 lo cual no es correcto

df[df['Exam_Score'] > 100][['Exam_Score']]
```

```python
df = df[df['Exam_Score'] <= 100]
```

### Visualización de resultados

```python
# Histogramas variables Numericas

for col_type in df.select_dtypes(include = "number").columns:
    sns.histplot(data = df, x = col_type, bins=20)
    plt.show()
```

```python
#Grafico Pastel Variables Cualitativas


for col_type in df.select_dtypes(include="object").columns:
    category_counts = df[col_type].value_counts()

    if category_counts.empty:
        print(f"Column '{col_type}' has no non-NaN values. Skipping pie chart.")
        continue 
  
    plt.pie(category_counts, labels=category_counts.index, autopct="%1.1f%%")  

   
    plt.title(f"Distribuicion de '{col_type}'") 
    plt.axis("equal")  
    plt.show()
```

```python
bins = np.array([val for val in range(50,105,5)])

plt.hist(df.Previous_Scores,bins,histtype='step',label='Previous Scores')
plt.hist(df.Exam_Score,bins,label="Final Exam Score")
plt.xlabel("MARKS")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

```python
# Diagrama de dispersión entre los examenes y precias calificaciones
sns.scatterplot(x='Exam_Score', y='Previous_Scores', data=df)
plt.show()
```

```python
# Matriz de correlación
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
